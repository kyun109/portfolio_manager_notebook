# Task 1 - Price Change

Suppose you buy a stock for one price, and sell it at the **new** price.

**Fill in** the function `price_change(old_price, new_price)` that returns the difference between the new price and the old price.

Example:
`price_change(10, 15)` returns `5`

If the price decreased, the result should be negative.

**To run the code, hover over the code block and press the PLAY button on the left**  
🢇

```python
def price_change(old_price, new_price):
    # Write your code here
    change = new_price - old_price
    return change

# Runs your code, try putting different values into it
print( price_change(10, 15) )

# vvvvv You will see the result printed below vvvvv
```

# Task 4: Yes or No?
**Create a function** `made_profit(old_price, new_price)` that returns a **Boolean** `True` if the `new_price` is greater, else returns `False`

Example: `made_profit(50, 85)` returns `True`.  
## Explanation: *Booleans*  
Booleans are a core **data type** (kind of value) present in almost **every programming language**.  
A *boolean* can only have **2** values, `True` or `False`.  

A boolean can be directly declared, like `my_bool = True`,  
**OR**  
They can be obtained from **comparing values**.  
`my_bool = 5 > 8` returns `False`

```python
# Write your function here
def made_profit(old_price, new_price):
    if new_price > old_price:
        return True
    else:
        return False

# This will run your function
print( made_profit(50, 85) )
```

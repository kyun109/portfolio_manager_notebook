# Task 2 - Average Opening Price

Sometimes it's useful to know the **average** price of a stock over several days.

**Create a function** `average_open_price(prices)` that returns the average value of all prices in the list.

Example:

`average_open_price([10, 20, 30])` returns `20.0`

```python
def average_open_price(prices):
    adding = 0
    for numbers in prices:
        adding = adding + numbers
    mean = adding / len(prices)
    return mean

print(average_open_price([105.20, 104.85, 106.40, 103.75]))
```
